```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# --- Configuration for Debugging Mode ---
DEBUG_SINGLE_IMAGE_MODE = True  # Set to True for single image debugging with plots
                                # Set to False for iterative processing without extra plots

# If DEBUG_SINGLE_IMAGE_MODE is True, specify the single image path
if DEBUG_SINGLE_IMAGE_MODE:
    single_image_path = "C:/Users/varou/OneDrive - UWE Bristol/Resit ItMV/Project Files/detection/detection/test/images/dataset1_front_31.png"
# ----------------------------------------

# Base directory where your images are located (used in iterative mode)
image_directory = "C:/Users/varou/OneDrive - UWE Bristol/Resit ItMV/Project Files/detection/detection/test/images/"

def process_image(image_path, debug_mode=False):
    """
    Processes a single image for apple detection.
    debug_mode: If True, prints/shows all intermediary steps.
    """
    # Load image
    image = cv2.imread(image_path)

    if image is None:
        print(f"Error: Could not load image from {image_path}")
        return None, 0

    if not debug_mode:
        print(f"Processing image: {os.path.basename(image_path)}")

    # Convert image to HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Updated red HSV range (these ranges might need fine-tuning for your specific images)
    lower_red1 = np.array([0, 80, 50])
    upper_red1 = np.array([20, 255, 255])
    lower_red2 = np.array([160, 80, 50])
    upper_red2 = np.array([180, 255, 255])

    # Create and combine red masks
    mask1 = cv2.inRange(hsv_image, lower_red1, upper_red1)
    mask2 = cv2.inRange(hsv_image, lower_red2, upper_red2)
    red_mask = cv2.add(mask1, mask2)

    # Morphological Operations for Watershed Prep
    kernel = np.ones((3, 3), np.uint8)

    # Noise removal with opening
    opening = cv2.morphologyEx(red_mask, cv2.MORPH_OPEN, kernel, iterations=2)

    # Sure background area
    sure_bg = cv2.dilate(opening, kernel, iterations=2)

    # Finding sure foreground area using Distance Transform
    dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
    ret, sure_fg = cv2.threshold(dist_transform, 0.2 * dist_transform.max(), 255, 0)

    # Unknown region (sure_bg - sure_fg)
    sure_fg = np.uint8(sure_fg)
    unknown = cv2.subtract(sure_bg, sure_fg)

    # Marker labelling
    ret, markers = cv2.connectedComponents(sure_fg)

    # Add one to all labels so that sure background is not 0, but 1
    markers = markers + 1

    # Mark the unknown region with 0
    markers[unknown == 255] = 0

    # Apply Watershed
    markers = cv2.watershed(image, markers)
    image[markers == -1] = [0, 0, 255]  # Mark watershed boundaries in red

    # Find contours from the watershed segmented image
    apple_count = 0
    contour_image = image.copy()

    for i in np.unique(markers):
        if i == -1 or i == 1:
            continue

        mask = np.zeros(image.shape[:2], dtype="uint8")
        mask[markers == i] = 255

        contours, _ = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        if contours:
            largest_contour = max(contours, key=cv2.contourArea)
            area = cv2.contourArea(largest_contour)

            min_apple_area = 10
            max_apple_area = 200000

            if min_apple_area < area < max_apple_area:
                perimeter = cv2.arcLength(largest_contour, True)
                if perimeter == 0:
                    continue

                circularity = 4 * np.pi * area / (perimeter**2)
                hull = cv2.convexHull(largest_contour)
                hull_area = cv2.contourArea(hull)
                solidity = area / hull_area if hull_area > 0 else 0

                if circularity > 0.2 and solidity > 0.2:
                    cv2.drawContours(contour_image, [largest_contour], -1, (0, 255, 0), 2)
                    apple_count += 1

    if debug_mode:
        # Display all intermediary results for debugging
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        contour_image_rgb = cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB)

        plt.figure(figsize=(20, 20))

        plt.subplot(2, 3, 1)
        plt.imshow(red_mask, cmap='gray')
        plt.title("Initial Red Mask")
        plt.axis('off')

        plt.subplot(2, 3, 2)
        plt.imshow(opening, cmap='gray')
        plt.title("After Opening (Noise Removal)")
        plt.axis('off')

        plt.subplot(2, 3, 3)
        plt.imshow(sure_bg, cmap='gray')
        plt.title("Sure Background")
        plt.axis('off')

        plt.subplot(2, 3, 4)
        plt.imshow(sure_fg, cmap='gray')
        plt.title("Sure Foreground (Markers)")
        plt.axis('off')

        plt.subplot(2, 3, 5)
        plt.imshow(unknown, cmap='gray')
        plt.title("Unknown Region")
        plt.axis('off')

        plt.subplot(2, 3, 6)
        plt.imshow(contour_image_rgb)
        plt.title(f"Detected Apples: {apple_count}")
        plt.axis('off')

        plt.show()

    return contour_image, apple_count

# --- Main execution logic based on the debug flag ---
if DEBUG_SINGLE_IMAGE_MODE:
    print("--- Running in Single Image Debugging Mode ---")
    _, final_apple_count = process_image(single_image_path, debug_mode=True)
    print(f"\nFinal apple count for {os.path.basename(single_image_path)}: {final_apple_count}")
else:
    print("--- Running in Iterative Processing Mode ---")
    # Iterate through files from 1 to 1000, adding 30 each step
    for i in range(1, 1001, 30):
        image_path = os.path.join(image_directory, f"dataset1_back_{i}.png")
        _, apple_count = process_image(image_path, debug_mode=False)
        if _ is not None: # Only print if image was loaded successfully
            print(f"Final apple count for {os.path.basename(image_path)}: {apple_count}")

        # You can optionally save the final detected image in iterative mode
        # cv2.imwrite(f"output_images/dataset1_back_{i}_detected.png", _)
```
