```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Image file path (replace with a path accessible to you if running locally)
image_path = "C:/Users/varou/OneDrive - UWE Bristol/Resit ItMV/Project Files/detection/detection/test/images/dataset1_back_1.png"

# Load image
image = cv2.imread(image_path)

if image is None:
    print(f"Error: Could not load image from {image_path}")
else:
    # Convert image to HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Define red color range in HSV (red wraps around hue)
    lower_red1 = np.array([0, 100, 100])
    upper_red1 = np.array([10, 255, 255])
    lower_red2 = np.array([170, 100, 100])
    upper_red2 = np.array([180, 255, 255])

    # Create masks for the two red ranges
    mask1 = cv2.inRange(hsv_image, lower_red1, upper_red1)
    mask2 = cv2.inRange(hsv_image, lower_red2, upper_red2)

    # Combine masks
    red_mask = cv2.add(mask1, mask2)

    # --- New: Apply Opening ---
    # Define a kernel for morphological operations.
    # A 3x3 or 5x5 kernel is a good starting point. Experiment with this size.
    kernel = np.ones((9, 9), np.uint8) # Increased kernel size slightly for potentially better noise removal

    # Apply Opening: Erosion followed by Dilation.
    # This removes small objects and noise, and can separate weakly connected objects.
    opened_mask = cv2.morphologyEx(red_mask, cv2.MORPH_OPEN, kernel)

    # --- Canny Edge Detection and Contour Finding ---
    # Now apply Canny edge detector to the OPENED mask
    # You might need to adjust the thresholds based on your image
    edges = cv2.Canny(opened_mask, 50, 150) # Use the opened_mask here

    # Find contours in the edges
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create a copy of the original image to draw contours on
    contour_image = image.copy()

    # Initialize a counter for potential apples
    apple_count = 0

    # Draw contours and filter based on area, circularity, and solidity
    for contour in contours:
        area = cv2.contourArea(contour)

        # Adjust min_area and max_area. Opening generally preserves object size better than pure erosion,
        # but you might still need slight adjustments.
        min_area = 80 # Example value, adjust as needed
        max_area = 100000 # Example value, adjust as needed

        if min_area < area < max_area:
            perimeter = cv2.arcLength(contour, True)
            if perimeter == 0:
                continue

            circularity = 4 * np.pi * area / (perimeter**2)
            hull = cv2.convexHull(contour)
            hull_area = cv2.contourArea(hull)
            solidity = area / hull_area if hull_area > 0 else 0

            # Adjust these thresholds based on how circular/solid your apples are
            min_circularity = 0.5 # Example, adjust
            min_solidity = 0.6   # Example, adjust

            if circularity > min_circularity and solidity > min_solidity:
                cv2.drawContours(contour_image, [contour], -1, (0, 255, 0), 2) # Green color, thickness 2
                apple_count += 1

    # Display results
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    masked_image_rgb = cv2.cvtColor(cv2.bitwise_and(image, image, mask=red_mask), cv2.COLOR_BGR2RGB) # Original masked image
    opened_mask_rgb = cv2.cvtColor(cv2.bitwise_and(image, image, mask=opened_mask), cv2.COLOR_BGR2RGB) # Opened masked image for visual
    edges_rgb = cv2.cvtColor(cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR), cv2.COLOR_BGR2RGB) # Convert grayscale edges to RGB for display
    contour_image_rgb = cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB)


    plt.figure(figsize=(25, 8))

    plt.subplot(1, 5, 1)
    plt.imshow(image_rgb)
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(1, 5, 2)
    plt.imshow(red_mask, cmap='gray')
    plt.title("Red Color Mask")
    plt.axis('off')

    plt.subplot(1, 5, 3)
    plt.imshow(opened_mask, cmap='gray') # Display the opened mask
    plt.title("Opened Red Mask")
    plt.axis('off')

    plt.subplot(1, 5, 4)
    plt.imshow(edges, cmap='gray')
    plt.title("Canny Edges on Opened Mask")
    plt.axis('off')

    plt.subplot(1, 5, 5)
    plt.imshow(contour_image_rgb)
    plt.title(f"Potential Apples Count: {apple_count}")
    plt.axis('off')

    plt.show()

    print(f"Rough count of potential apples: {apple_count}")
```
